```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X =[[2,1],[3,1],[1,4],[2,6]]
```

```python
# 归一化到[-1,1]
scaler = MinMaxScaler(feature_range=(-1,1))
# 将缩放器应用于特征矩阵X
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
from sklearn.preprocessing import StandardScaler
print(X)
```

```python
# 标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```
