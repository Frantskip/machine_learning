```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义函数f(x)
def J(x):
    return (x**2-2)**2

# 定义梯度函数（导数）
def gradient(x):
    return 4*x*(x**2-2)

# 用列表保存点的变化轨迹
x_list = []
y_list = []
```

```python
# 定义超参数和x的初始值
alpha = 0.01
x = 1

# 重复迭代100次
while np.abs(grad := gradient(x)) > 1e-10:
    y = J(x)
    x_list.append(x)
    y_list.append(y)
    print(f"x={x:.4f}, y={y:.4f}")

    # 更新x
    x = x - alpha * grad

print(len(x_list))
```

```python
# 可视化函数f(x)和梯度下降的过程
x = np.arange(0,2, 0.01)
plt.plot(x, J(x), label="J(x)")
plt.plot(x_list, y_list, "r")
plt.scatter(x_list, y_list, c="r")
plt.show()
```

```python
# 局部放大
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(x, J(x), label="J(x)")
ax[0].plot(x_list, y_list, "r")
ax[0].scatter(x_list, y_list, c="r")
ax[0].set_title("Global View")
ax[0].set_xlabel("x")
ax[0].set_ylabel("J(x)")
ax[0].legend()

x_list2 = x_list[1:]
y_list2 = y_list[1:]
x = np.arange(1.399,1.425, 0.001)
ax[1].plot(x, J(x), label="J(x)")
ax[1].plot(x_list2, y_list2, "r")
ax[1].scatter(x_list2, y_list2, c="r")
ax[1].set_title("Zoomed View")
ax[1].set_xlabel("x")
ax[1].set_ylabel("J(x)")
ax[1].legend()
```
