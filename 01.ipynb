```python
import numpy as np
```

向量

```python
x = np.array([1, 2, 3])
print(x)
print(type(x))
print(x.shape)
```

```python
print(x.T)
print(x.T.shape)
```

```python
# 计算范数
print(np.linalg.norm(x, ord=2))
```

```python
# 矩阵运算
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)
print(A.shape)
B = np.array([[1,3,7],[5,0,2]])
print(B)
print(B.shape)
```

```python
# 1.转置
print(A.T)
print(A.T.shape)
```

```python
# 2.矩阵相加
print(A + B)
```

```python
# 3.矩阵与标量相乘
print(A * 3)
```

```python
# 4.矩阵相乘
print(A @ B)
```
