```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X = [[2,1],[3,1],[1,4],[2,6]]
y = [0,1,0,1]
```

```python
knn = KNeighborsRegressor(n_neighbors=2)  # 定义KNN回归器，k=2
knn.fit(X, y)  # 训练模型
# 4. 预测新样本
x = [[4,9]]
x_pred = knn.predict(x)  # 预测新样本的回归值
print(x_pred)  # 输出新样本的回归值
```
