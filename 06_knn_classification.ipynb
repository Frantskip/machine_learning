```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 加载数据集
X = np.array([[2,1],[3,1],[1,4],[2,6]])
y = np.array([0,0,1,1])  # 分类标签
```

```python
# 定义KNN分类器
knn = KNeighborsClassifier(n_neighbors=2,weights="distance")  # 定义KNN分类器，k=2

# 2. 训练模型
knn.fit(X, y)  # 训练模型

# 3. 预测新样本
x =np.array([[4,9]])  
x_class = knn.predict(x)  # 预测新样本的分类
print(x_class)  # 输出新样本的分类
```

```python
# 画图
# 使用bool索引将数据分为两类
X1=X[y==0]
X2=X[y==1]
# 定义不同的颜色
colors = ["C0","C1"]
plt.scatter(X1[:,0],X1[:,1],c=colors[0])
plt.scatter(X2[:,0],X2[:,1],c=colors[1])
# 新点的颜色
x_color=colors[0] if x_class==0 else colors[1]
plt.scatter(x[:,0],x[:,1],c=x_color,marker="x")
plt.title("KNN Classification")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
